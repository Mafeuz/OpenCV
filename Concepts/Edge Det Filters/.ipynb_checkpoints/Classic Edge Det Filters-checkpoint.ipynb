### Sobel Filter:

* Apply a convolution operation with a matrix of coefs based on the pixels frequency gradient
* Select if it will look over X axis, Y axis or both
* Less affected by noise
* Not rotationally symmetric
* Can give information of angle and strength of a edge/border

```python
# Starting with conversion to gray-scale:
img_gray = cv2.cvtColor(img_in_use, cv2.COLOR_BGR2GRAY)

# Apply Blur for better results (REALLY IMPORTANT TO GET BETTER RESULTS):
img_gray = cv2.GaussianBlur(img_gray, (9,9), 3, 3)

img_sobel_X  = cv2.Sobel(src=img_gray, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=3)
img_sobel_Y  = cv2.Sobel(src=img_gray, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=3)
img_sobel_XY = cv2.Sobel(src=img_gray, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=3)

img_sobel_X  = cv2.cvtColor(img_sobel_X.astype('uint8'), cv2.COLOR_GRAY2RGB)
img_sobel_Y  = cv2.cvtColor(img_sobel_Y.astype('uint8'), cv2.COLOR_GRAY2RGB)
img_sobel_XY = cv2.cvtColor(img_sobel_XY.astype('uint8'), cv2.COLOR_GRAY2RGB)

OpCV_Utils.show_multiple_imgs_plt([[img_sobel_X, img_sobel_Y, img_sobel_XY]],
                                  [['Sobel X', 'Sobel Y', 'Sobel XY']])
```

```python
# Let's take X and Y separated to get better results as we add them on and check the result:

img_sobel_XY_added = cv2.addWeighted(img_sobel_X, 0.5, img_sobel_X, 0.5, 0)

OpCV_Utils.show_single_img_plt(img_sobel_XY_added, 'Sobel XY Added Result', (5,5))
```

### Scharr Filter
* Not rotationally symmetric
* More affected by noise

```python
# Starting with conversion to gray-scale:
img_gray = cv2.cvtColor(img_in_use, cv2.COLOR_BGR2GRAY)

# Apply Blur for better results:
img_gray = cv2.GaussianBlur(img_gray, (9,9), 3, 3)

img_scharr_X  = cv2.Scharr(src=img_gray, ddepth=cv2.CV_64F, dx=1, dy=0)
img_scharr_Y  = cv2.Scharr(src=img_gray, ddepth=cv2.CV_64F, dx=0, dy=1)
img_scharr_XY = cv2.addWeighted(img_scharr_X, 0.5, img_scharr_X, 0.5, 0)

img_scharr_X  = cv2.cvtColor(img_scharr_X.astype('uint8'), cv2.COLOR_GRAY2RGB)
img_scharr_Y  = cv2.cvtColor(img_scharr_Y.astype('uint8'), cv2.COLOR_GRAY2RGB)
img_scharr_XY = cv2.cvtColor(img_scharr_XY.astype('uint8'), cv2.COLOR_GRAY2RGB)

OpCV_Utils.show_multiple_imgs_plt([[img_scharr_X, img_scharr_Y, img_scharr_XY]],
                                  [['Scharr X', 'Scharr Y', 'Scharr XY']])
```

```python
# Let's Compare Scharr and Sobel:

OpCV_Utils.show_multiple_imgs_plt([[img_sobel_X, img_sobel_Y, img_sobel_XY_added],
                                   [img_scharr_X, img_scharr_Y, img_scharr_XY]],
                                  [['Sobel X', 'Sobel Y', 'Sobel XY'],
                                   ['Scharr X', 'Scharr Y', 'Scharr XY']], fig_size = (15,8),)
```

### Laplacian Filter:
* Second derivative filter (check gradient derivative zeros)
* More affected by noise
* Faster to Compute

```python
# Starting with conversion to gray-scale:
img_gray = cv2.cvtColor(img_in_use, cv2.COLOR_BGR2GRAY)

# Apply Blur for better results:
img_gray = cv2.GaussianBlur(img_gray, (9,9), 3, 3)

img_laplacian = cv2.Laplacian(img_gray, cv2.CV_64F, ksize=3)

img_laplacian  = cv2.cvtColor(img_laplacian.astype('uint8'), cv2.COLOR_GRAY2RGB)
OpCV_Utils.show_single_img_plt(img_laplacian, 'Laplacian - Kernel = (3,3)', fig_size=(9,9))
```
