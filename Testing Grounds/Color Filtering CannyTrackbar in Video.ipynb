```python
import numpy as np
import cv2
import time
from OpCV_Utils import Display_Multiple_Images, color_filtering, custom_canny
```

```python
# Load Video:
video = 'ztest_media/people.mp4'

cap = cv2.VideoCapture(video)
video_fps = cap.get(cv2.CAP_PROP_FPS)

############################################################################
#trackbar callback function does nothing but required for trackbar
def nothing(x):
    pass

# Filter Controls Trackbar:
cv2.namedWindow('Filter Controls')
cv2.createTrackbar('R min','Filter Controls', 0,255, nothing)
cv2.createTrackbar('G min','Filter Controls', 0,255, nothing)
cv2.createTrackbar('B min','Filter Controls', 0,255, nothing)

cv2.createTrackbar('R max','Filter Controls', 0,255, nothing)
cv2.createTrackbar('G max','Filter Controls', 0,255, nothing)
cv2.createTrackbar('B max','Filter Controls', 0,255, nothing)

# Canny Controls Trackbar:
cv2.namedWindow('Canny Controls')
cv2.createTrackbar('Dilation','Canny Controls', 0,10, nothing)
cv2.createTrackbar('Erosion','Canny Controls', 0,10, nothing)
cv2.createTrackbar('Canny Thresh X','Canny Controls', 10,255, nothing)
cv2.createTrackbar('Canny Thresh Y','Canny Controls', 10,255, nothing)
############################################################################

while True:
    
    start = time.time()
    
    success, frame = cap.read()
    
    if not success:
        break
        
    frame = cv2.resize(frame, (600,500))
    
    # Keyboard Controls:
    
    key = cv2.waitKey(1) or 0xff   
        
    if key == ord('k'):
        break
    
    ############################################################################
    # Color Filtering with Trackbar:
    
    R_min = int(cv2.getTrackbarPos('R min','Filter Controls'))
    G_min = int(cv2.getTrackbarPos('G min','Filter Controls'))
    B_min = int(cv2.getTrackbarPos('B min','Filter Controls'))
    
    R_max = int(cv2.getTrackbarPos('R max','Filter Controls'))
    G_max = int(cv2.getTrackbarPos('G max','Filter Controls'))
    B_max = int(cv2.getTrackbarPos('B max','Filter Controls'))
    
    # define color boundaries to filtering highlight (BGR)
    color_filter_boundaries = [([B_min, G_min, R_min], [B_max, G_max, R_max])]
    
    frame_filtered = color_filtering(frame, color_filter_boundaries)
    
    dilation = int(cv2.getTrackbarPos('Dilation','Canny Controls'))
    erosion = int(cv2.getTrackbarPos('Erosion','Canny Controls'))
    cny_thresh = (int(cv2.getTrackbarPos('Canny Thresh X','Canny Controls')), 
                  int(cv2.getTrackbarPos('Canny Thresh Y','Canny Controls')))
    
    frame_canny = custom_canny(frame, kernel_size = (2,2), canny_thresh = cny_thresh, 
                               dil_level = dilation, ero_level = erosion)
    
    frame = Display_Multiple_Images([[frame, frame_filtered], [frame_canny]], scale = 0.5)
    
    ############################################################################
    
    end = time.time()

    frame_time = (end - start) + 0.0001
    fps = np.floor(1/frame_time)
        
    if (fps > video_fps):
        time.sleep(1/video_fps)
        fps = video_fps
        time.sleep(0.1)
            
    cv2.putText(frame, "FPS: {}".format(fps), (30,20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)

    cv2.imshow("Output", frame)
    
cv2.destroyAllWindows()
cap.release()

print('R min:', R_min)
print('G min:', G_min)
print('B min:', B_min)
print('R max:', R_max)
print('G max:', G_max)
print('B max:', B_max)

print('Dilation Iterations:', dilation)
print('Erosion Iterations:', erosion)
print('Canny Thresh:', cny_thresh)
```
