```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import time
```

```python
import sys
sys.path.append('C:/Users/Mafeus/Desktop/Git_Repos/OpenCV')
import OpCV_Utils
```

```python
img = cv2.imread('img1.png')
wheel = cv2.imread('wheel.png')
angle = 0

dx = 0
dy = 0

xt = -1
yt = -1

while True:
    
    img_use = img.copy()
    
    angle += 5
    dx    += 6*xt
    dy    += 3*yt
        
    rotated = OpCV_Utils.img_rotation(wheel, angle, pivot=(wheel.shape[1]//2, wheel.shape[0]//2), keep_full_img=False)
        
    mask = np.zeros_like(img)
    
    mask[:rotated.shape[0], :rotated.shape[1], :] = rotated
    
    # Move to desired position:
    
    cX, cY = (300,200)
    
    mask = OpCV_Utils.img_translation(mask, cX+dx, cY+dy)
        
    img_use[np.where((mask != [0,0,0]).all(axis = 2))] = [0,0,0]
    
    output = cv2.bitwise_or(img_use, mask)
    
    mag, phase = OpCV_Utils.imgFourrierTransform(output)
    
    output = cv2.resize(output, (480, 360))
    mag = cv2.resize(mag, (480, 360))

    if (angle > 360):
        angle = 0
    
     # Keyboard Controls:
    
    key = cv2.waitKey(1) or 0xff   
        
    if key == ord('k'):
        break
    
    stack = OpCV_Utils.stackImgs([[mag, mask]], sep_lines=True, scale=1)
    cv2.imshow('rotation', stack)
    
    if (abs(dx)>200):
        xt = xt*(-1)
        
    if (abs(dy)>200):
        yt = yt*(-1)
        
cv2.destroyAllWindows() 
```

```python
########################################### Video Stream Loop #######################################################
#####################################################################################################################

# Load Video:
video_path = 'C:/Users/Mafeus/Desktop/Git_Repos/OpenCV/Testing Grounds/ztest_media/people.mp4'
cap = cv2.VideoCapture(video_path)
video_fps = cap.get(cv2.CAP_PROP_FPS)
frame_res = (600,500)

while True:
    
    ###############################################################################################################
    ###############################################################################################################
    # Get Frame:
    start = time.time()
    success, frame = cap.read()
    
    if not success:
        print('Server OFF')
        break
        
    frame = cv2.resize(frame, frame_res)
    
    ###############################################################################################################
    ############################################# Processing ######################################################
    mag, _ = OpCV_Utils.imgFourrierTransform(frame)
            
    frame = OpCV_Utils.stackImgs([[frame, mag]], sep_lines=True, scale=1)
    
    ###############################################################################################################
    # Display FPS:
    end = time.time()

    frame_time = (end - start) + 0.0001
    fps = np.floor(1/frame_time)
        
    if (fps > video_fps):
        time.sleep(1/video_fps)
        fps = video_fps
            
    cv2.putText(frame, "FPS: {}".format(fps), (30,20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)
    ###############################################################################################################
    # Keyboard Controls:
    key = cv2.waitKey(1) or 0xff   
    if key == ord('k'):
        break
    ###############################################################################################################
    # Show Frame:
    cv2.imshow("Video Stream", frame)
    ###############################################################################################################
    
cv2.destroyAllWindows()
cap.release()
```
