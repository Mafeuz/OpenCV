### OpenCV + YOLOv4(coco) for Object Detection ###

```python
import numpy as np
import time
import cv2
import matplotlib.pyplot as plt
```

```python
# Get coco class labels:

coco_labels = open('YOLO/YOLO/cfg/coco.names').read().strip().split('\n')
print(coco_labels)
print(len(coco_labels))
```

```python
# Get Yolo Net Architecture Weights and Configuration:
weights_path = 'YOLO/YOLO/yolov4.weights'
config_path = 'YOLO/YOLO/cfg/yolov4.cfg'

# Read cv2 net:
net = cv2.dnn.readNet(config_path, weights_path)

print(net)
```

```python
# Generate random colors for bboxes:
colors = np.random.randint(0, 255, size=(len(coco_labels), 3), dtype='uint8')
```

```python
# Check structure:
ln = net.getLayerNames()
print('Total of Layers: ' + str(len(ln)))

print('Output Layers:', net.getUnconnectedOutLayers())
```

```python
# Check unconnected layers:
ln = [ln[i[0] - 1] for i in net.getUnconnectedOutLayers()]
ln
```

```python
# Display img from cv2 format func:
def img_show(img):
    fig = plt.gcf()
    fig.set_size_inches(10,7)
    plt.axis('off')
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
# Check test img:
img = cv2.imread('YOLO/YOLO/imagens/cachorros.jpg')
img_show(img)
```

```python
# Make img copy for later use:
img_cp = img.copy()
```

```python
print(img_cp.shape)
(H, W) = img.shape[:2]
print('Height: ' + str(H) + '\nWidth: ' + str(W))
```

```python
# Pass img to Yolo and get output:

start= time.time()

# Modify img to input format:
blob = cv2.dnn.blobFromImage(img, 1 / 255.0, (416, 416), swapRB = True, crop = False)

# Pass to net:
net.setInput(blob)
layer_outputs = net.forward(ln)

end = time.time()
print('YOLO took {:.2f} seconds to process.'.format(end - start))
```

```python
len(layer_outputs)
```

```python
# Define necessary variables:
threshold = 0.5
threshold_NMS = 0.3
boxes = []
confiability = []
IDclasses = []
```

```python
# Checking Results (layer outputs):

for output in layer_outputs:
    for detection in output:
        scores = detection[5:]
        classeID = np.argmax(scores)
        trust = scores[classeID]
        if trust > threshold:
            print('scores: ' + str(scores))
            print('Probable Class: ' + str(classeID))
            print('Trust: ' + str(trust))

            box = detection[0:4] * np.array([W, H, W, H])
            (centerX, centerY, width, height) = box.astype('int')

            x = int(centerX - (width / 2))
            y = int(centerY - (height / 2))

            boxes.append([x, y, int(width), int(height)])
            confidence.append(float(trust))
            IDclasses.append(classeID)
```

```python
print(IDclasses), len(IDclasses)
```

```python
# Filtering detected object boxes:
objs = cv2.dnn.NMSBoxes(boxes, confidence, threshold, threshold_NMS)
print(objs.flatten())
```

```python
confidence[0], confidence[7], confidence[15], confidence[11]
```

```python
# Let's Draw bboxes for each detected object:

if len(objs) > 0:
    for i in objs.flatten():
        (x, y) = (boxes[i][0], boxes[i][1])
        (w, h) = (boxes[i][2], boxes[i][3])

        objct = img_cp[y:y + h, x:x + w]
        img_show(objct)
       
        color = [int(c) for c in colors[IDclasses[i]]]

        cv2.rectangle(img, (x, y), (x + w, y + h), color, 2)
        text = "{}: {:.4f}".format(coco_labels[IDclasses[i]], confidence[i])
        cv2.putText(img, text, (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

img_show(img)
```

```python
cv2.imwrite('result.jpg', img)
```
