```python
import numpy as np
import cv2
import time
import matplotlib.pyplot as plt
import face_recognition as f_rec
```

```python
import sys
sys.path.append('C:/Users/Mafeus/Desktop/Git_Repos/OpenCV')
import OpCV_Utils
```

```python
willsmith1 = f_rec.load_image_file('media/will_smith1.jpg')
willsmith2 = f_rec.load_image_file('media/will_smith2.jpg')
willsmith3 = f_rec.load_image_file('media/will_smith3.jpg')

willemdafoe1 = f_rec.load_image_file('media/willem_dafoe1.jpg')
willemdafoe2 = f_rec.load_image_file('media/willem_dafoe2.jpeg')
willemdafoe3 = f_rec.load_image_file('media/willem_dafoe3.jpg')
```

```python
OpCV_Utils.show_multiple_imgs_plt([[willsmith1, willsmith2, willsmith3]],
                                  [['Will Smith 1', 'Will Smith 2', 'Will Smith 3']], fig_size = (10,3))
```

```python
OpCV_Utils.show_multiple_imgs_plt([[willemdafoe1, willemdafoe2, willemdafoe3]],
                                  [['Willem Dafoe 1', 'Willem Dafoe 2', 'Willem Dafoe 3']], fig_size = (10,3))
```

```python
# Detecting Faces and get the face encondings:
faceLoc1 = f_rec.face_locations(willsmith1)[0]
encodedWillSmith1 = f_rec.face_encodings(willsmith1)[0]
copy1 = willsmith1.copy()
cv2.rectangle(copy1, (faceLoc1[3], faceLoc1[0]), (faceLoc1[1], faceLoc1[2]), (255,0,0), 5)

faceLoc2 = f_rec.face_locations(willsmith2)[0]
encodedWillSmith2 = f_rec.face_encodings(willsmith2)[0]
copy2 = willsmith2.copy()
cv2.rectangle(copy2, (faceLoc2[3], faceLoc2[0]), (faceLoc2[1], faceLoc2[2]), (255,0,0), 20)

faceLoc3 = f_rec.face_locations(willsmith3)[0]
encodedWillSmith3 = f_rec.face_encodings(willsmith3)[0]
copy3 = willsmith3.copy()
cv2.rectangle(copy3, (faceLoc3[3], faceLoc3[0]), (faceLoc3[1], faceLoc3[2]), (255,0,0), 5)
```

```python
OpCV_Utils.show_multiple_imgs_plt([[copy1], [copy2], [copy3]],
                                  [['Will Smith 1'], ['Will Smith 2'], ['Will Smith 3']], fig_size = (15,15))
```

```python
# Checking if it matches:
results = f_rec.compare_faces([encodedWillSmith1], encodedWillSmith2)
faceDis = f_rec.face_distance([encodedWillSmith1], encodedWillSmith2)

print(results, faceDis)

results = f_rec.compare_faces([encodedWillSmith1], encodedWillSmith3)
faceDis = f_rec.face_distance([encodedWillSmith1], encodedWillSmith3)

print(results, faceDis)
```

```python
# Detecting Faces and get the face encondings:
faceLoc1 = f_rec.face_locations(willemdafoe1)[0]
encodedwillemdafoe1 = f_rec.face_encodings(willemdafoe1)[0]
copy1 = willemdafoe1.copy()
cv2.rectangle(copy1, (faceLoc1[3], faceLoc1[0]), (faceLoc1[1], faceLoc1[2]), (255,0,0), 5)

faceLoc2 = f_rec.face_locations(willemdafoe2)[0]
encodedwillemdafoe2 = f_rec.face_encodings(willemdafoe2)[0]
copy2 = willemdafoe2.copy()
cv2.rectangle(copy2, (faceLoc2[3], faceLoc2[0]), (faceLoc2[1], faceLoc2[2]), (255,0,0), 5)

faceLoc3 = f_rec.face_locations(willemdafoe3)[0]
encodedwillemdafoe3 = f_rec.face_encodings(willemdafoe3)[0]
copy3 = willemdafoe3.copy()
cv2.rectangle(copy3, (faceLoc3[3], faceLoc3[0]), (faceLoc3[1], faceLoc3[2]), (255,0,0), 5)
```

```python
OpCV_Utils.show_multiple_imgs_plt([[copy1], [copy2], [copy3]],
                                  [['Willem Dafoe 1'], ['Willem Dafoe 2'], ['Will Dafoe 3']], fig_size = (15,15))
```

```python
# Checking if it matches:
results = f_rec.compare_faces([encodedwillemdafoe1], encodedwillemdafoe2)
faceDis = f_rec.face_distance([encodedwillemdafoe1], encodedwillemdafoe2)

print(results, faceDis)

results = f_rec.compare_faces([encodedwillemdafoe1], encodedwillemdafoe3)
faceDis = f_rec.face_distance([encodedwillemdafoe1], encodedwillemdafoe3)

print(results, faceDis)
```
